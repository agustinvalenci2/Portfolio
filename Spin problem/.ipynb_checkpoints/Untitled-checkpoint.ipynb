```python
import networkx as nx
import random
```

```python
m=10500
random.seed(0)
tree = nx.random_tree(n=m, seed=0)
```

```python
z=random.randint(5,10)
print("c iadjaio")
print("p test09",m,m+z)
for u,v in tree.edges:
    print(u,v,random.randint(100,5000))
for i in range(z+1):
    r=random.randint(0,m-1)
    print(r,r,random.randint(-5000,5000))
```
